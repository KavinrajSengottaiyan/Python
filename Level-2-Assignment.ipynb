```python
# print Odd Numbers in the list
tup3=(20,10,16,19,25,1,276,188)
print(tup3)
for num in tup3:
    if(num%2==1):
        print(num, "is odd")
```

```python
# print Even Numbers in the list
print(tup3)
for num in tup3:
    if(num%2==0):
        print(num, "is even")
```
