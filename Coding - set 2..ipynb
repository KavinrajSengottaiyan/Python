```python
#Write a Python program to generate the Fibonacci sequence up to 10 terms.
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
    for i in fib(n):
        print(fib(i), end=" ")
```

```python
fib(10)
```

```python
#Write a Python function to check if a given string is a palindrome.

pal=input("Enter")

if pal==pal[::-1]:
    print(pal,"is a palindrome")
else:
    print(pal,"is not a palindrome")
```

```python
#Write a Python program to count the number of vowels in a given string. 

def vowels(text):
    vow = "aeiou"
    return sum(1 for char in text.lower() if char in vow)
text=input("Enter text: ")
print("No. of Vowels:", vowels(text))
```
